# Introduction to Neural Networks

## Operation

```python
class Operation():
    
    def __init__(self, input_nodes=[]):
        
        self.input_nodes = input_nodes
        self.output_nodes = []
        
        for node in input_nodes:
            node.output_nodes.append(self)
            
        _default_graph.operations.append(self)
            
        def compute(self):
            pass
```

### Addition

```python
class add(Operation):
    
    def __init__(self, x, y):
        super().__init__([x, y])
        
    def compute(self, x_var, y_var):
        self.inputs = [x_var, y_var]
        return x_var + y_var
```

### Multiplication

```python
class multiply(Operation):
    
    def __init__(self, x, y):
        super().__init__([x, y])
        
    def compute(self, x_var, y_var):
        self.inputs = [x_var, y_var]
        return x_var * y_var
```

### Matrix Multiplication

```python
class matmul(Operation):
    
    def __init__(self, x, y):
        super().__init__([x, y])
        
    def compute(self, x_var, y_var):
        self.inputs = [x_var, y_var]
        # We assume that x_var and y_var are numpy matrices and we can perform dot product on them
        return x_var.dot(y_var)
```

<b>Placeholder</b> - An "empty" node that needs a value to be provided to compute output <br>
<b>Variable</b> - Changeable parameter of Graph <br>
<b>Graph</b> - Global Variable connecting variables and placeholders to operations

## Placeholders

```python
class Placeholder():
    
    def __init__(self):
        
        self.output_nodes = []
        
        _default_graph.placeholders.append(self)
```

## Variables

```python
class Variable():
    
    def __init__(self, initial_value = None):
        
        self.value = initial_value
        self.output_nodes = []
        
         
        _default_graph.variables.append(self)
```

## Graph

```python
class Graph():
    
    def __init__(self):
        
        self.operations = []
        self.placeholders = []
        self.variables = []
        
    def set_as_default(self):
        
        global _default_graph
        _default_graph = self
```

## Traversing Operation Nodes

```python
def traverse_postorder(operation):
    """ 
    PostOrder Traversal of Nodes. Basically makes sure computations are done in 
    the correct order (Ax first , then Ax + b). Feel free to copy and paste this code.
    It is not super important for understanding the basic fundamentals of deep learning.
    """
    
    nodes_postorder = []
    def recurse(node):
        if isinstance(node, Operation):
            for input_node in node.input_nodes:
                recurse(input_node)
        nodes_postorder.append(node)

    recurse(operation)
    return nodes_postorder
```

## Session

```python
import numpy as np
```

```python
class Session:
    
    def run(self, operation, feed_dict = {}):
 
        # Puts nodes in correct order
        nodes_postorder = traverse_postorder(operation)
        
        for node in nodes_postorder:

            if type(node) == Placeholder:
                
                node.output = feed_dict[node]
                
            elif type(node) == Variable:
                
                node.output = node.value
                
            else: # Operation
                
                node.inputs = [input_node.output for input_node in node.input_nodes]

                 
                node.output = node.compute(*node.inputs)
                
            # Convert lists to numpy arrays
            if type(node.output) == list:
                node.output = np.array(node.output)
        
        # Return the requested node value
        return operation.output



        return operation.output
```

```python
sess = Session()
```

```python
result = sess.run(operation=z, feed_dict={x:10})
```

## Activation Function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
sample_z = np.linspace(-10, 10, 100)
sample_a = sigmoid(sample_z)
```

```python
plt.plot(sample_z, sample_a)
```

```python
class Sigmoid(Operation):
    
    def __init__(self,z):
        super().__init__([z])
            
    def compute(self, z_val):
        return 1 / (1 + np.exp(-z_val))
```
